```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [8000, 500, 80, 155, 107, 1314, 261, 2187] #all
# x = [8000, 500, 80, 155, 107, 1314, 37, 805] #attached
y = [30, .3, .06, .1, 0.065, 1.29, 0.066, 2.64]
times = pd.DataFrame(np.transpose([x, y]), columns=["x", "y"]).sort_values("x", ignore_index=True)
print(times)
coefs = np.polyfit(x, y, 2)
print(coefs)
poly = np.poly1d(coefs)
print(poly)
times2 = times.copy()
times2.y = np.sqrt(times.y)
```

```python
# times.plot()
plt.plot(range(0,8000,10),[poly(x) for x in range(0,8000,10)] )
plt.plot(times.x, times.y)
```
